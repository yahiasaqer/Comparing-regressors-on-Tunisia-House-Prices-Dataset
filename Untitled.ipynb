```python
#importing pandas and numpy
import pandas as pd
import numpy as np
prices = pd.read_csv("dataSetFull.csv")
prices
```

```python
#here we see if there are any null values.
missing = pd.DataFrame({'missing': prices.isnull().sum()})
missing
```

```python
(prices==-1).sum()
```

```python
#here we fill null values using the mean for the "Area" and "price_tnd" columns 

prices['Area'] = prices['Area'].replace(np.nan, prices["Area"].mean())

prices['price_tnd'] = prices['price_tnd'].replace(np.nan, prices["price_tnd"].mean())


#we replace the "age" values with numeric values that represent age catogeries
prices.age.replace(('0','1-5 ','5-10 ','10-20 ','10,20','20-30 ', '30-50 ', '50-70 ', '70-100 ','Plus de 100 '),(1,2,3,4,4,5,6,7,8,9), inplace=True)
prices["age"].head()

#now we replace null values with the mean, then we convert the columns values back to integer, because it does not make
#sense to have 3.5 rooms for example




prices['pieces'] = prices['pieces'].replace(np.nan, prices["pieces"].mean())
pieces_int = prices['pieces'].astype(int)
prices['pieces'] = prices['pieces'].replace(prices["pieces"].values, pieces_int)

prices['room'] = prices['room'].replace(np.nan, prices["room"].mean())
room_int = prices['room'].astype(int)
prices['room'] = prices['room'].replace(prices["room"].values, room_int)

prices['bathroom'] = prices['bathroom'].replace(np.nan, prices["bathroom"].mean())
bath_int = prices['bathroom'].astype(int)
prices['bathroom'] = prices['bathroom'].replace(prices["bathroom"].values, bath_int)



prices['age'] = prices['age'].replace(np.nan, prices["age"].mean())
age_int = prices['age'].astype(int)
prices['age'] = prices['age'].replace(prices["age"].values, age_int)

prices['state'] = prices['state'].replace(np.nan, prices["state"].mean())
state_int = prices['state'].astype(int)
prices['state'] = prices['state'].replace(prices["state"].values, state_int)

prices['latt'] = prices['latt'].replace(np.nan, prices["latt"].mean())
latt_int = prices['latt'].astype(int)
prices['latt'] = prices['latt'].replace(prices["latt"].values, latt_int)

prices['long'] = prices['long'].replace(np.nan, prices["long"].mean())
long_int = prices['long'].astype(int)
prices['long'] = prices['long'].replace(prices["long"].values, long_int)

prices['distance_to_capital'] = prices['distance_to_capital'].replace(np.nan, prices["distance_to_capital"].mean())
distance_to_capital_int = prices['distance_to_capital'].astype(int)
prices['distance_to_capital'] = prices['distance_to_capital'].replace(prices["long"].values, distance_to_capital_int)

missing = pd.DataFrame({'missing': prices.isnull().sum()})
missing
```

```python
#we use the boxplot function to plot the data to see if there are any outliers
import matplotlib.pyplot as plt
plt.figure(figsize= (20, 10))
prices.boxplot()
plt.show()

#and it  seems that we have a lot of outliers, anyways, we are not going to do anything about them.
```

```python
#now we check the correlation, in order to select the features the we will use for our regression tasks in respect to the "price_tnd" feature
prices.corr(method='pearson')
```

```python
prices.corr(method = "kendall")
```

```python
prices.corr(method = "spearman")
```

```python
#from the data above, and after using 3 correlation analysis methods, we conclude that 'Area', "room", "bathroom" and "pieces"
# have the highest values when it comes to correlation in respect to the "price_tnd"
x = np.array(prices[['Area', "room", "bathroom", "pieces"]])
y = np.array(prices['price_tnd'])

#here we split that data to testing and training sets, 30% of the data is selected for testing, the rest (70%) is the training data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3 )
```

```python
#firstly we use the linear regression.
from sklearn.linear_model import LinearRegression
model = LinearRegression()
#train the model...
model.fit(X_train, y_train)
```

```python
#now we test the model
y_pred = model.predict(X_test)
```

```python
#printing the coefficients (slopes) and the Y intercept
print(model.coef_[0], "*Area + ",model.coef_[1], "*room + ",model.coef_[2], "*bathroom + ",model.coef_[3], "*pieces + ", model.intercept_  )
```

```python
# importing the math library to use the square root
import math  
#importing mean_squared_error
from sklearn.metrics import mean_squared_error
#now we calculate the RMSE
mse =  mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)
print("The mean squared error is: ", mse, ", while the root mean squared error is: ", rmse, ".")
```

```python
# we repeat the same steps for the following 3 regressors
from sklearn.svm import SVR
model = SVR()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mse =  mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)
print("The mean squared error is: ", mse, ", while the root mean squared error is: ", rmse, ".")
```

```python
from sklearn.ensemble import RandomForestRegressor
# number of trees is 100
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mse =  mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)
print("The mean squared error is: ", mse, ", while the root mean squared error is: ", rmse, ".")
```

```python
#importing "curve_fit" from scipy
from scipy.optimize import curve_fit

# define the exponential (nonlinear) function 
def func(x, a, b, c):
    return a* x**b+c 
```

```python
#choosing the "Area" since it has the highest value using spearman and kendall coeffecint analysis
x2 = prices["Area"][0:1000]
y2 = prices['price_tnd'][0:1000]
#number of iterations is 10000
popt, pcov = curve_fit(func, x2, y2,  maxfev=10000)
print(popt)
y2pred = func(x2, *popt)
```

```python
mse =  mean_squared_error(y2,y2pred)
rmse = math.sqrt(mse)
print("The mean squared error is: ", mse, ", while the root mean squared error is: ", rmse, ".")
```

```python
#defining a linear function to use for fit_curve regressor
def func(x, a, b):
    return a*x + b
#number of iterations is 800
popt, pcov = curve_fit(func, x2, y2)
print(popt)
y2pred = func(x2, *popt)
```

```python
mse =  mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)
print("The mean squared error is: ", mse, ", while the root mean squared error is: ", rmse, ".")
```
